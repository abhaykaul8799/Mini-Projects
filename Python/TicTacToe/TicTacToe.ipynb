```python
'''
Python code for TicTacToe game.
'''

from IPython.display import clear_output
from time import sleep
from sys import exit

board=[' ',' ',' ',' ',' ',' ',' ',' ',' ']

def printboard(board):
    '''
    Prints the board on the screen after clearing the previous output
    '''
    clear_output()
    print(board[6] + '|' + board[7] + '|' + board[8])
    print(board[3] + '|' + board[4] + '|' + board[5])
    print(board[0] + '|' + board[1] + '|' + board[2])
  

def edit(board,marker,pos):
    '''
    Edit the list for the corresponding marker
    '''
    if pos>=0 and pos<9:
        if board[pos] == ' ':
            board[pos] = marker
            printboard(board)
            return True
        else:
            print("Invalid Position!")
            return False
    else:
        print("Invalid Position!")
        return False

def checkwin(board):
    '''
    Check for the win conditions
    '''
    if board[0]==board[1]==board[2]=='X' or board[0]==board[1]==board[2]=='O':
        return board[0]
    elif board[3]==board[4]==board[5]=='X' or board[3]==board[4]==board[5]=='O':
        return board[3]
    elif board[6]==board[7]==board[8]=='X' or board[6]==board[7]==board[8]=='O':
        return board[6]
    elif board[0]==board[3]==board[6]=='X' or board[0]==board[3]==board[6]=='O':
        return board[0]
    elif board[1]==board[4]==board[7]=='X' or board[1]==board[4]==board[7]=='O':
        return board[1]
    elif board[2]==board[5]==board[8]=='X' or board[2]==board[5]==board[8]=='O':
        return board[2]
    elif board[0]==board[4]==board[8]=='X' or board[0]==board[4]==board[8]=='O':
        return board[0]
    elif board[2]==board[4]==board[6]=='X' or board[2]==board[4]==board[6]=='O':
        return board[2]
    else:
        return False

p1marker=input("Enter marker for Player 1 (X or O): ")
p1marker=p1marker.upper()
while p1marker != 'X' and p1marker != 'O':
    print("Invalid Input!")
    p1marker = input("Enter marker for Player 1 (X or O): ")
    p1marker = p1marker.upper()
clear_output()
if p1marker == 'X':
    p2marker = 'O'
else:
    p2marker = 'X' 
print("Marker for player 1 is: " + p1marker)
print("Marker for player 2 is: " + p2marker)

m=[] #marker list
p=[] #player list
x=int(input("Which player will move first? (1/2): "))
while x!=1 and x!=2:
    print("Invalid Input!")
    x=int(input("Which player will move first? (1/2): "))
if x == 1:
    m.append(p1marker)
    m.append(p2marker)
    p=["Player 1","Player 2"]
elif x == 2:
    m.append(p2marker)
    m.append(p1marker)
    p=["Player 2","Player 1"]

count=0
while checkwin(board) == False:
    pos=int(input("{}, enter the position where you will place your marker '{}'. (1-9) ".format(p[count%2],m[count%2])))
    while pos<0 or pos>9:
        print("Invalid Output!")
        pos=int(input("{}, enter the position where you will place your marker '{}'. (1-9) ".format(p[count%2],m[count%2])))
    if(edit(board,m[count%2],pos-1)==False):
        count-=1
    count+=1
    if board.count('X')+board.count('O')==9:
        printboard(board)
        if checkwin(board):
            break
        print("Draw!!!")
        sleep(10)        
        exit()


winner = checkwin(board)
if p1marker == winner:
    print("Player 1 Wins!!!")
    sleep(10)
elif p2marker == winner:
    print("Player 2 Wins!!!")
    sleep(10)
```
